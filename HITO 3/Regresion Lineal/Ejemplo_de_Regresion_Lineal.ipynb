### Regresion con statsmodels
y = beta0 + beta1*x1 + beta2*x2 + ... + betan*xn

```python
import statsmodels.formula.api as snf #Para hacer la regresion
```

```python
# Regresion Simple (agregamos una sola X)
ln = snf.ols(formula="precio~superficie_util", data=data).fit()
```

```python
ln.summary()
# y = 182200 + 136.98 * superficie_util
```

```python

pred = ln.predict(data['superficie_util'])
```

```python
plt.scatter(data.superficie_util, data.precio, color='red',s=3)
plt.scatter(data.superficie_util, pred, color='black', s=2)
plt.title('Datos reales Vs. Estimación')
plt.xlabel('superficie util')
plt.ylabel('precio en UF')
plt.show()
##nube de puntos --> datos reales
##linea negra --> estimación
```

```python
# Regresion Lineal Multiple: agregamos mas de un X
lm2 = snf.ols(formula="precio~superficie_util+banos+dormitorios+anios", data=data).fit()
```

```python
lm2.summary()
```

### Regresion lineal con Scikit Learn

```python
from sklearn.feature_selection import RFE  # Nos ayuda con la seleccion de variables
#Para la estimación
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

```python
feature_cols = ['superficie_total', 'superficie_util', 'dormitorios', 'banos', 'anios', 'grado']
```

```python
x = data[feature_cols]
y = data['precio']
```

```python
estimator = SVR(kernel="linear")

selector = RFE(estimator, n_features_to_select=2, step=1)
selector = selector.fit(x,y)
```

```python
selector.support_
```

```python
list(zip(feature_cols, selector.support_))
```

```python
selector.ranking_
```

```python
x_pred = x[['banos','grado']]
```

```python
lm3 = LinearRegression()
lm3.fit(x_pred,y)
```

```python
lm3.intercept_
```

```python
lm3.coef_
```

```python
list(zip(x_pred, lm3.coef_))
```

##### precio = -249594.9716975948 + banos * 19539.958383101843  + grado * 87038.6863416177

```python
lm3.score(x_pred, y)
```
